```python
import util

import jax
import jax.numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
def get_weekly(samples, start, obs_names=['mean_y', 'mean_z'], **kwargs):
    
    fields = {}
    for obs_name in obs_names:
        if obs_name in samples:
                obs0 = samples[obs_name + '0'][:, None]
                obs = samples[obs_name]
                obs = np.concatenate((obs0, obs), axis=1)
                if obs_name + '_future'  in samples:
                    obs_future = samples[obs_name + '_future']
                    obs = np.concatenate((obs, obs_future), axis=1)
                fields[obs_name] = obs.astype('float32')

    n_samples, T = fields[obs_names[0]].shape

    t = pd.date_range(start=start, periods=T, freq='D')

    weekly_dfs = []
    
    for obs_name in obs_names:
        A = fields[obs_name].T # the numpy array
        weekly_df = pd.DataFrame(index=t, data=A).resample("1w").last().diff().shift(-1).iloc[:-1]
        weekly_dfs.append(weekly_df)
        
    return weekly_dfs
```
